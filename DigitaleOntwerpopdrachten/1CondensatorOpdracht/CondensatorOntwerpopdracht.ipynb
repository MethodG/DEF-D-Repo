##### Vraag 1a Het potentiaal rond een puntlading

1. Bepaal welke index correspondeert met $z = 0$ m.
2. Plot de log van het potentiaal tegen de log van de straal.

````python
# 1
print('1: Er is geen waarde voor z die precies 0 is, z[50] geeft de kleinste waarde:', -np.min(np.abs(z)))
index_z0 = np.argmin(np.abs(z))  # index van de z-waarde die het dichtst bij 0 ligt

# 2 & 3
V_slice = V[index_z0, :]  

def fit_1a3(x, a, b):
    return x**a * b

x_1a3 = np.linspace(np.min(r[1:]), np.max(r), 1000)  # we negeren het eerste punt (r=0) omdat de potentiaal daar oneindig is

popt_1a3, pcov_1a3 = curve_fit(fit_1a3, r[1:], np.abs(V_slice[1:]))  # we negeren het eerste punt (r=0) omdat de potentiaal daar oneindig is

plt.figure(figsize=(10, 8))
plt.plot(r, V_slice, 'bo', label='Data (V vs r)')
plt.plot(x_1a3, fit_1a3(x_1a3, *popt_1a3), 'r--', label=f'Fit: V ~ r^{popt_1a3[0]:.2f}')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('Radius (m)')
plt.ylabel('Absolute waarde van de potentiaal')
plt.title('Elektrisch potentiaal \n(log-log plot)')
plt.legend()
plt.show()

# 4
print(f'4: De exponent a is ongeveer {popt_1a3[0]:.2f}, wat dicht bij -1 ligt, wat overeenkomt met de theoretische verwachting voor een puntlading (V ~ 1/r).')
````
